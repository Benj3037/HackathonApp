```python
#Import libraries
import altair as alt
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import pandas as pd
```

```python
# Directory containing the main folder
main_dir = "put the path where you have unzipped the files"
```

```python
# Get list of subfolders
subfolders = [f.name for f in os.scandir(main_dir) if f.is_dir()]

# Dictionary to hold your dataframes
dfs = {}

for subfolder in subfolders:
    # Concatenate main folder with subfolder name
    folder_path = os.path.join(main_dir, subfolder)
    
    # Get the single CSV file in the subfolder
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    # Assuming there's only one CSV file, fetch it
    if len(csv_files) == 1:
        csv_file = csv_files[0]
        # Load the data from CSV
        df = pd.read_csv(os.path.join(folder_path, csv_file))
        
        # Create a dynamic variable name based on subfolder name
        var_name = "df_" + subfolder[:3]
        dfs[var_name] = df
    else:
        print(f"More than one CSV file found in {subfolder}, or none found.")
```

```python
for k in dfs:
    print(k)
```

```python
def display_info(df):
    """
    Display basic information about the DataFrame.
    """
    print("\n--- Info ---\n")
    print(df.info())
```

```python
def display_head(df, n=5):
    """
    Display the first n rows of the DataFrame.
    """
    print("\n--- Head ---\n")
    print(df.head(n))
```

```python
def display_describe(df):
    """
    Display descriptive statistics of the DataFrame.
    """
    print("\n--- Describe ---\n")
    print(df.describe())
```

```python
def display_single(df):
    """
    For a single DataFrame, display its info, head, and descriptive statistics.
    """
    display_info(df)
    display_head(df)
    display_describe(df)
    print("\n" + "-"*50 + "\n")
```

```python
# Now you can use:
for k in dfs:
    print(f"DataFrame: {k}")
    display_single(dfs[k])
```
